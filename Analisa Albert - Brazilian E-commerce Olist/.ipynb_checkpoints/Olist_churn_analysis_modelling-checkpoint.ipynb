Looking at the distribution of review scores, the most actually review 1. So the most comebacks are unhappy about the order. 

There are approximately 3 times more orders which are non-comeback orders. Therefore a balancing of the data will be made. The following code is taking from the course materials but has been customized to work on the Olist data:


```python
################## CODE INSPIRTED BY Part 1 - Spam Classification-solutions by *Stanislav Borysov [stabo@dtu.dk], DTU Management* ##########

# balancing example: upsampling the minor class (spam)
from sklearn.utils import resample

def balance_data(df):
    # Separate majority and minority classes
    df_comeback = df[df.Comeback==1]
    df_leave = df[df.Comeback==0]

    # Upsample minority class
    df_comeback_upsampled = resample(df_comeback, replace=True, n_samples=len(df_leave),
                                 random_state=42)

    # Combine majority class with upsampled minority class
    df_balanced = pd.concat([df_leave, df_comeback_upsampled])

    # Display new class counts
    print(df_balanced.Comeback.value_counts())
    
    return df_balanced
```

Now saving the balanced data with an equal proportion of both non-comebacks and come-backs:

```python
all_persons_first_order_2016_2017_balanced=balance_data(all_persons_first_order_2016_2017)
```

```python
# REMOVE EMOJIS lav et for-loop

import re

no_emoji_text = []
emojitext = data_final['review_comment_message']

for text in emojitext:
  emoji_pattern = re.compile("["
          u"\U0001F600-\U0001F64F"  # emoticons
          u"\U0001F300-\U0001F5FF"  # symbols & pictographs
          u"\U0001F680-\U0001F6FF"  # transport & map symbols
          u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                            "]+", flags=re.UNICODE)

  no_emoji_text.append((emoji_pattern.sub(r'', text)))

data_final['No_emoji'] = no_emoji_text
```

```python
stopwordsPT = nltk.corpus.stopwords.words('portuguese') 
```

```python
stopwordsPT[100:110] # cheking to see if it worked
```

```python
def text_processing(text):
      
    # remove punctuation 
    text = "".join([c for c in text 
                   if c not in string.punctuation])
    
    # lowercase
    text = "".join([c.lower() for c in text])
    # remove stopwords
    text = " ".join([w for w in text.split() 
                     if w not in stopwordsPT])
    # stemming / lematizing (optional)
    text = " ".join([stemmer.stem(w) for w in text.split()]) # removed to better understand translated words in topics
    return text
```

```python
data_final['message_clean'] = data_final['No_emoji'].apply(text_processing)
```

###Purpose of doing topic modelling using LDA <br>
We will do topic modelling using Latent Dirichlet Allocation.
Topic modelling finds a number of latent topics in natural language data, that can be used amongst other things for dimensionality reduction. In our case we have a vocabulary size of 3151, so the term-frequency representations will have a dimension of 3151. As we wish to compare and make sense of values of variable coefficients, plugging in 3151 variables to represent the review comments in a logistic regression model is not a feasible option. Instead, LDA can help us reduce the dimensionality of the representation of the review comments, while hopefully keeping as much as possible of the explanatory power.<br>

###Chosing value of hyperparameter n_topics <br>
Before doing LDA, we need to find the number of topics we should use for the LDA. So we split the training data so it also has a validation set in it, and do grid search to tune the hyperparameter of number of topics for LDA:

```python
# Split into train/validate

rs = 42 # reproducible results, set to None for random
(X_len_train_sub, X_len_val_sub, 
 X_bow_counts_train_sub, X_bow_counts_val_sub, 
 X_bow_tfidf_train_sub, X_bow_tfidf_val_sub, 
 y_train_sub, y_val_sub) = train_test_split(X_len_train, X_bow_counts_train, X_bow_tfidf_train, y_train,test_size=0.2, stratify=y_train, random_state=rs)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report


n_topics_list = list(range(1, 40, 1)) # (1,1000,20) max iter 50: omkring 0.65 | (1,200,10) max iter 100: ditto 
performance = []
#
for n_topics in n_topics_list:
    lda = LatentDirichletAllocation(n_components=n_topics, max_iter=100,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)
    X_lda_train_sub = lda.fit_transform(X_bow_counts_train_sub)
    X_lda_val_sub = lda.transform(X_bow_counts_val_sub)
    parameters = {'C': [10e-10, 0.1, 1, 10, 10e10]}
    lr = GridSearchCV(LogisticRegression(class_weight='balanced', solver='newton-cg'), 
                      parameters, cv=3)
    lr.fit(X_lda_train_sub, y_train_sub)
    print(lr.best_params_)
    y_pred = lr.predict(X_lda_val_sub)
    report = classification_report(y_val_sub, y_pred, 
                                   target_names=labels, output_dict=True)
    performance.append(report['macro avg']['f1-score'])
```

```python
plt.plot(n_topics_list, performance)
plt.show();
```

On the plot we see that the macro avg f1-score rises to around 0.61 at 5 topics, and then doesn't meaningfully improve. Since we want to interpret on the actual meaning of each topic, we wish to use a smaller number of topics rather than a larger number. So we choose to do the LDA topic modelling with 5 topics. 

###Creating LDA 
We create and train the model:

```python
from sklearn.decomposition import LatentDirichletAllocation

n_topics = 5 # this a hyper-parameter from cv above
lda = LatentDirichletAllocation(n_components=n_topics, max_iter=100,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)
```

```python
X_lda_train = lda.fit_transform(X_bow_counts_train)
```

### Printing topics and buidling wordclouds
We print, save and translate the words and weights inside each topic for purposes of visualisation. The weights are also normalised, so the biggest weight of each topic is equal to 1. These weights are not used for modelling, but only for sensemaking through the word clouds. 

```python
def print_top_words(model, feature_names, n_top_words):
    norm = model.components_.sum(axis=1)[:, np.newaxis]
    for topic_idx, topic in enumerate(model.components_):
        print(80 * "-")
        print("Topic {}".format(topic_idx))
        for i in topic.argsort()[:-n_top_words - 1:-1]:
            print("{:.3f}".format(topic[i] / norm[topic_idx][0])
                  + '\t' + feature_names[i]
                  + '\t' + translator.translate(feature_names[i])) # translates to english
```

```python
print("\nTopics in LDA model:")
counts_feature_names = count_vect.get_feature_names()
n_top_words = 15 # not hyper-param, just for printing
print_top_words(lda, counts_feature_names, n_top_words)
```

One of the assumptions of logistic regression is the absence of multicollinearity among the predictors. So to be able to test the impact on multicollinearity on the model, and to remove the least explanatory of the found topics, we extract each individual topic, and run a logistic regression explaining churn with all topics, and then each individual topic:

```python

# extract individual topic values to avoid multicollinearity

X_lda_train_0 = []
X_lda_train_1 = []
X_lda_train_2 = []
X_lda_train_3 = []
X_lda_train_4 = []

for i in range(len(X_lda_train)):
  X_lda_train_0.append(X_lda_train[i][0])
  X_lda_train_1.append(X_lda_train[i][1])
  X_lda_train_2.append(X_lda_train[i][2])
  X_lda_train_3.append(X_lda_train[i][3])
  X_lda_train_4.append(X_lda_train[i][4])


X_lda_test_0 = []
X_lda_test_1 = []
X_lda_test_2 = []
X_lda_test_3 = []
X_lda_test_4 = []

for i in range(len(X_lda_test)):
  X_lda_test_0.append(X_lda_test[i][0])
  X_lda_test_1.append(X_lda_test[i][1])
  X_lda_test_2.append(X_lda_test[i][2])
  X_lda_test_3.append(X_lda_test[i][3])
  X_lda_test_4.append(X_lda_test[i][4])

```

**Comments on coefficients with significant impacts and derived insights**<br>
<br>
*Price*<br>
The higher the price the lower the probability of returning. So marketing products with a lower price to new customers might be a good way to heighten the probability of the customer returning. In the long run it may pay off, since multiple items of lower value could results in higher revenue than fewer items of higher value.<br>
<br>
*Product name length*<br>
This might be a case of a spurious correlation. In the sense that people with a general tendency to purchase many items might be the type of people who also like long product names. But the correlation between longer product names of chosen products and higher probability of returning are both only really impacted by the tendency to buy many items - so the tendency to buy many items is the confounding variable.<br>
<br>
*Product description length*<br>
It seems that a shorter description heightens the probability of returning. But the magnitude of the coefficient is small and the theoretical explanation: "I would like to purchase another item from Olist, since the description of the item I have already purchased was nice and short", doesn't make much sense.<br>
<br>
*Topic 0: belated delivery*<br>

# Deep learning - building a Keras neural network for classification

## Building a Keras Feed Forward Neural Network using numerical order metrics

**Introduction** <br>
Logistic regression has already been tried out to classify whether an order will be the order of a customer who will churn. For purposes of improving the classification of churn, a non-linear model will be tried out. This model is a Feed Forward Neural Network (FFNN).  

**The problem**<br>
The target values, comebacks or not (the y's), are known in the classification problem. The machine learning is therfore supervised, when preassuming what the model should predict. 

**Hidden layers** <br>
The hidden layers of the network are used to increase the the non-linearity and change the representation of data for better generalization.

**Output layer**<br>
The classification problem is a binary problem, and the desired output of the last layer in the network is to classify using order characteritics whether or not a customer will churn (1 for return or 0 for churn). Therefore, a sigmoid activation function is used in the output layer, because it's output is between 0 and 1. If the sigmoid's output is larger than 0.5, it outputs 1. If the sigmoid's output is less than 0.5, it outputs 0. 

**Data** <br>
To improve the classification of come-back customers a Keras model will be build. The first Keras model uses the following numerical variables:


*   Price 
*   Freight value
*   Review score
*   Product name length
*   Product description length
*   Product photos qty
*   product_weight_g
*   product_length_cm
*   product_height_cm
*   product_width_cm
*   Volume (length X height X width)



```python
from keras.layers import Dense
from keras.models import Sequential
from keras.layers import Dropout
```

Defining the Keras model:

```python
model = Sequential()
```

Choosing 100 neurons for the hidden layers. The number of hidden neurons is a hyperparameter, and after testing several values for the number of neurons, 100 performs the best. The number of classes are equal to two (returning customers and churning customers)

```python
hidden_neurons_n = 100
num_classes = 2
```

Six hidden layers has been implemented to the model. The choice of activation in the hidden layers is tanh. The properties of tanh is the same as sigmoid, but it is also continous and differentiable at all points - therefore it can  easily be used in back-propagation. Dropout between the first and the second layer has been introduced for the purpose of improving the generalization of the model.

```python
#Hidden layers
model.add(Dense(hidden_neurons_n))
model.add(Dropout(0.4))
model.add(Dense(hidden_neurons_n, activation='tanh'))
model.add(Dense(hidden_neurons_n, activation='tanh'))
model.add(Dense(hidden_neurons_n, activation='tanh'))
model.add(Dense(hidden_neurons_n, activation='tanh'))
model.add(Dense(hidden_neurons_n, activation='tanh')) 
model.add(Dense(hidden_neurons_n, activation='tanh'))

#Output layer
model.add(Dense(1, activation='sigmoid'))
```

### F1-score for Keras

To compare the accuracy of the text analysis classifiers, the logistic regression and the FFNN a f1-score function will be used. The function has been found on https://medium.com/@aakashgoel12/how-to-add-user-defined-function-get-f1-score-in-keras-metrics-3013f979ce0d 

```python
import keras.backend as K

def get_f1(y_true, y_pred): #taken from old keras source code
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision+recall+K.epsilon())
    return f1_val
```

**Discussion**<br>
When looking at the results of the logistic regression model 3 out of 4 topics were insignificant and the direction of the impact on churn of the significant topic was different to what we theoretically would have expected. But when using the same topics in the feed forward neural network, performance jumped from 0.69 to 0.82. This could be because the logistic regression model lacks the ability to model more complex and latent structures, that might have revealed themselves more when used in a feed forward neural network.<br>
<br> 
Furthermore, when using the sklearn logistic regression implementation, we were able to include all 5 topics (because of the default regularisation). But when using statsmodels this wasn't possible (due to multicollinearity of the topics, since they sum to 1 for each review message).<br> 
<br> 
All this plays into the general distinction between statistics and machine learning, or "explainability" vs. "accuracy". In this project we show both sides of the coin, so to speak, in that we could find some, but not a lot of theoretically meaningful effects of variables on the probability of churning, while using the same variables in a neural network created a sizable jump in model performance - but also a different interpretation of the impact of the topics on prediction, since they went from almost no explanatory power in logistic regression to being high impact variables in the neural networks. This doesn't mean that one approach to modelling is better than the other, but we find it interesting that the topics found with LDA could have so radically different impacts on prediction beween the two modelling paradigms. It also shows the tradeoffs made in this regard. When performing logistic regression with statsmodels, we get comparable coefficients within a 95 pct. confidence interval and p-values that show the significance of the variables. When building neural networks we have to sacrifice some of this information, but in this case we were rewarded with extra explanatory power, that could not be extracted from the topics using a more traditional and less complex modelling approach. <br>
<br>
Finally, we showed that the logistic regression model, contrary to its reputation in this case, was highly affectable by a small subset of outliers. Furthermore, the removal of some (theoretically) irrelevant variables - in our case product name length - had a very significant impact on other variables. This is a useful result for Olist, in the sense that when doing similar analysis in the future, they should be aware of these unexpected relations in the dataset, that do not have clear theoretical explanations.
